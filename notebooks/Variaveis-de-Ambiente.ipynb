# Variáveis de Ambiente #

Variáveis de ambiente são valores armazenados fora do código da aplicação e carregados durante a execução do sistema, elas servem para armazenar informações de configuração como:
Endereço do banco de dados
Porta do banco
Usuário
Senha
Chaves de API
Configurações de ambiente

Inicialmente a conexão estava fixa no código:

DATABASE_URL = (
    "postgresql://postgres:postgres@localhost:5432/usuarios_db"
)

**Problemas:**

Senha exposta
Dificuldade para mudar de ambiente
Necessidade de alterar código para trocar configurações

**Solução**

Utilizar um arquivo .env.

**Arquivo .env**

```python
DB_HOST=localhost
DB_PORT=5432
DB_USER=postgres
DB_PASSWORD=postgres
DB_NAME=usuarios_db
```

**Carregando as variáveis**

```python
from dotenv import load_dotenv
import os

load_dotenv()

DB_HOST = os.getenv("DB_HOST")
DB_PORT = os.getenv("DB_PORT")
DB_USER = os.getenv("DB_USER")
DB_PASSWORD = os.getenv("DB_PASSWORD")
DB_NAME = os.getenv("DB_NAME")
```

**Montando a URL de conexão**

```python
DATABASE_URL = (
    f"postgresql://{DB_USER}:{DB_PASSWORD}"
    f"@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)
```

**Resultado**

postgresql://postgres:postgres@localhost:5432/usuarios_db

# Benefícios #

**Segurança**

A senha não fica exposta no código.

**Flexibilidade**

Desenvolvimento:

DB_HOST=localhost

Produção:

DB_HOST=10.0.0.15

Container:

DB_HOST=db

Nenhuma alteração no código é necessária.
