# Pool de Conexões # 

Um pool de conexões é um conjunto de conexões abertas com o banco de dados que ficam disponíveis para reutilização pela aplicação.

Sem pool, a cada requisição a API precisa:

Abrir conexão
↓
Executar consulta
↓
Fechar conexão

Com um pool:

Pool
├── Conexão 1
├── Conexão 2
├── Conexão 3
├── Conexão 4
└── Conexão 5

A API apenas pega uma conexão disponível e a devolve quando termina de utilizá-la.

# Configuração do Pool #

engine = create_engine(
    DATABASE_URL,
    pool_size=5,
    max_overflow=10,
    pool_timeout=30,
    pool_recycle=1800,
    pool_pre_ping=True,
)

pool_size: Mantém X conexões abertas permanentemente.
max_overflow: Permite criar conexões extras quando todas as conexões do pool estiverem ocupadas.
pool_timeout: Tempo máximo de espera por uma conexão livre, se todas estiverem ocupadas: Esperar até X segundos, caso nenhuma seja liberada: TimeoutError
pool_recycle: Recicla conexões antigas após X segundos.

Conexão antiga
↓
Descartada
↓
Nova conexão criada

pool_pre_ping: Antes de entregar uma conexão ao sistema, o SQLAlchemy verifica se ela ainda está ativa.

Conexão válida?
↓
Sim → Utiliza
Não → Cria nova conexão


# Fluxo de Funcionamento #


Cliente
↓
FastAPI
↓
get_db()
↓
Pool entrega conexão
↓
Consulta PostgreSQL
↓
Commit ou Rollback
↓
db.close()
↓
Conexão retorna ao Pool

OBS: db.close(): Não fecha a conexão física, a conexão apenas retorna ao pool para reutilização.
